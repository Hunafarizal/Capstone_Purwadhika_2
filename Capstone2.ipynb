# 2. DATA CLEANING

Sekarang kita akan memasuki tahap kedua yaitu *Data cleaning*. 

Tahap ini sangatlah krusial dalam analisis data yang bertujuan untuk meningkatkan kualitas data dengan menghilangkan ketidakakuratan, kesalahan, atau inkonsistensi. Proses ini sangat penting karena kualitas data yang tinggi adalah prasyarat utama untuk mendapatkan wawasan yang akurat dan berharga dari analisis data. Berikut adalah beberapa alasan mengapa tahap pembersihan data sangat penting dalam analisis data:

### 1. **Meningkatkan Akurasi Hasil Analisis**
Data yang tidak akurat atau penuh dengan kesalahan dapat menyebabkan analisis yang salah arah atau kesimpulan yang keliru. Misalnya, data duplikat, nilai yang hilang, atau nilai yang salah dapat mengganggu analisis statistik, yang pada gilirannya mempengaruhi keputusan bisnis atau penelitian yang diambil berdasarkan data tersebut.

### 2. **Memastikan Konsistensi**
Dalam kumpulan data besar, terutama yang dikumpulkan dari sumber yang berbeda, sering terjadi inkonsistensi. Misalnya, variasi dalam format tanggal, kesalahan ejaan dalam data kategorikal, atau penggunaan unit ukuran yang berbeda dapat menyebabkan analisis yang tidak konsisten. Pembersihan data membantu menyatukan standar dan format, sehingga semua data dapat diintegrasikan dan dianalisis dengan benar.

### 3. **Menghilangkan Duplikasi**
Duplikasi data sering terjadi karena kesalahan dalam pengumpulan data atau sebagai akibat dari penggabungan dataset dari beberapa sumber. Data duplikat bisa memberikan representasi yang salah tentang data dan mempengaruhi analisis, seperti menghitung rata-rata atau median. Mengidentifikasi dan menghapus entri duplikat adalah langkah penting untuk memastikan integritas analisis.

### 4. **Menangani Data yang Hilang**
Data yang hilang bisa terjadi karena berbagai alasan, termasuk kesalahan dalam pengumpulan data atau masalah dalam transmisi data. Teknik pembersihan data dapat mencakup imputasi (mengisi nilai yang hilang dengan estimasi berdasarkan data lain) atau penghapusan baris/kolom dengan terlalu banyak nilai yang hilang, tergantung pada jumlah dan alasan ketiadaan data tersebut.

### 5. **Optimalisasi Penggunaan Sumber Daya**
Data yang bersih memungkinkan proses analisis yang lebih cepat dan efisien. Data yang tidak terstruktur atau penuh dengan kesalahan dapat memperlambat proses komputasi dan menyebabkan penggunaan sumber daya yang tidak efisien, baik dari segi waktu maupun biaya komputasi.

### 6. **Memfasilitasi Integrasi Data**
Dalam banyak kasus, data diperoleh dari berbagai sumber yang mungkin memiliki format yang berbeda-beda. Pembersihan data memungkinkan integrasi yang mulus dari berbagai sumber data, yang penting untuk analisis komprehensif dan multidimensional.

### 7. **Meningkatkan Kepercayaan Stakeholder**
Data yang bersih dan akurat meningkatkan kepercayaan pemangku kepentingan dalam analisis yang dilakukan. Kepercayaan ini penting untuk mendukung keputusan berbasis data di semua tingkatan organisasi atau dalam konteks penelitian.

Karena semua alasan ini, tahap pembersihan data dianggap sebagai salah satu bagian yang paling penting dalam proses analisis data. Tanpa pembersihan yang efektif, risiko membuat keputusan yang tidak optimal atau salah sangat meningkat.

```python
# Membuat salinan dari data original
original_data = transjakarta_data.copy()
```

Sekarang mari kita konversikan terlebih dahulu Kolom yang tidak sesuai deskripsi dengan tipe data nya dan juga perihal `direction` yang akan kita replace keterangan nya demi mempermudah pengertian

```python
# Modifikasi date and time columns ke datetime objects
original_data['tapInTime'] = pd.to_datetime(original_data['tapInTime'])
original_data['tapOutTime'] = pd.to_datetime(original_data['tapOutTime'])

# Modifikasi keterangan pada kolom 'direction'
original_data['direction'] = original_data['direction'].replace({0: 'Keberangkatan', 1: 'Kedatangan'})
```

Sekarang kita mulai proses Data Cleaning nya. Kita akan mulai dengan mengisi missing values pada `corridorID` dan `corridorName`

Karena keduanya memiliki hubungan yang erat, kita akan menggunakan data yang ada untuk mengisi nilai yang hilang pada salah satu kolom berdasarkan kolom yang lain. Jika `CorridorName` hilang, kita akan mencoba mengisinya berdasarkan nilai yang sesuai dari `CorridorID`, begitu juga sebaliknya.


```python
# Pertama, kita akan membuat pemetaan dari data yang tersedia untuk CorridorID dan CorridorName.
corridor_id_to_name = original_data.dropna(subset=['corridorID', 'corridorName']).drop_duplicates('corridorID').set_index('corridorID')['corridorName'].to_dict()
corridor_name_to_id = original_data.dropna(subset=['corridorID', 'corridorName']).drop_duplicates('corridorName').set_index('corridorName')['corridorID'].to_dict()
```

```python
# Memperbaiki case sensitive huruf nama kolom dan mencoba mengisi lagi
original_data['corridorName'] = original_data.apply(
    lambda row: corridor_id_to_name.get(row['corridorID'], row['corridorName']) if pd.isna(row['corridorName']) else row['corridorName'],
    axis=1
)

# Verifikasi koreksi dan periksa nilai yang hilang untuk kolom-kolom ini setelah diisi
original_data[['corridorID', 'corridorName']].isna().sum()
```

Kesulitan dalam pengisian ini mungkin karena beberapa  `CorridorID` tidak memiliki pasangan `CorridorName` yang sesuai atau sebaliknya dalam data yang tersedia.

Untuk `corridorID` dan `corridorName` hanya jika kedua dua nya sama-sama kosong, kita akan isi missing values nya dengan keterangan 'Tidak Tersedia'



```python
# Fill missing values dengan 'Tidak Tersedia' untuk kedua 'corridorID' dan 'corridorName' yang dimana kedua-duanya kosong
mask = original_data['corridorID'].isna() & original_data['corridorName'].isna()
original_data.loc[mask, ['corridorID', 'corridorName']] = 'Tidak Tersedia'

# Verifikasi apakah pengisian berhasil dan periksa nilai yang masih hilang dalam kolom-kolom ini
original_data[['corridorID', 'corridorName']].isna().sum()
```

Berhubungan `corridorID` dengan corridorName saling berkaitan, dan juga dikarenakan `corridorName` sudah sepenuhnya bersih dari missing values, maka kita dapat mengisi sisa `corridorID` dengan duplikat keterangan dari kolom `corridorName`

```python
# Bangun kembali pemetaan dari corridorName ke corridorID menggunakan data lengkap yang ada.
updated_corridor_name_to_id = original_data.dropna(subset=['corridorID', 'corridorName']).drop_duplicates('corridorName').set_index('corridorName')['corridorID'].to_dict()

# Isi 'corridorID' yang hilang berdasarkan 'corridorName'
original_data['corridorID'] = original_data.apply(
    lambda row: updated_corridor_name_to_id.get(row['corridorName'], row['corridorID']) if pd.isna(row['corridorID']) else row['corridorID'],
    axis=1
)

original_data['corridorID'].isna().sum()
```

Sekarang `corridorID` dan `corridorName` sudah tidak ada missing Values lagi.

Selanjutnya, mari kita lanjutkan dengan mengisi nilai-nilai yang hilang pada kolom-kolom yang berkaitan dengan `tapOutStops`, `tapOutStopsName`, `tapOutStopsLat`, `tapOutStopsLon`, `stopEndSeq`, dan `tapOutTime`. Untuk ini, kita akan mencari dan menggunakan baris data lengkap sebagai referensi untuk mengisi nilai yang hilang di baris lain yang memiliki kecocokan dalam beberapa kolom kunci.

Proses ini akan dilakukan dengan mencari data yang serupa atau cocok dan menggunakan nilai dari data tersebut untuk mengisi kekosongan.

Kita akan mencoba mengisi nilai-nilai yang hilang berdasarkan asumsi bahwa nilai yang hilang terjadi secara bersamaan pada semua kolom ini, yang berarti jika salah satu kolom ini hilang pada suatu baris, semua kolom lainnya di baris yang sama mungkin juga hilang. Mari kita lanjutkan.


```python
# Group related columns for tapOut information
tapout_columns = ['tapOutStops', 'tapOutStopsName', 'tapOutStopsLat', 'tapOutStopsLon', 'stopEndSeq', 'tapOutTime']

# Create a complete case subframe to use for imputation
complete_tapout = original_data.dropna(subset=tapout_columns)

# Define a function to fill missing values based on available complete cases
def fill_missing_tapout(row, reference_df, columns):
    if pd.isna(row['tapOutStops']):
        similar_rows = reference_df[reference_df['tapInStops'] == row['tapInStops']]
        if not similar_rows.empty:
            for col in columns:
                row[col] = similar_rows[col].iloc[0]
    return row

# Apply the function to fill missing values
original_data = original_data.apply(fill_missing_tapout, axis=1, args=(complete_tapout, tapout_columns))

# Check the remaining missing values for these tapOut columns after filling
original_data[tapout_columns].isna().sum()
```

Cleaning ini berhasil mengurangi jumlah missing values dengan cukup signifikan, tetapi masih ada beberapa baris yang tidak memiliki informasi lengkap yang cukup untuk dilakukan cleaning secara otomatis. Hal ini mungkin disebabkan oleh keunikan dari data tersebut yang tidak memiliki kemiripan dalam data yang sudah lengkap.

Maka dari kita kita akan melanjutkan pengisian missing values dengan menggunakan estimasi berdasarkan kolom lain yang berkaitan, seperti menggunakan rata-rata atau median dari nilai-nilai pada kolom yang sama.



Untuk mengisi nilai yang masih hilang pada kolom-kolom yang berkaitan dengan tapOut, kita akan menggunakan median untuk kolom numerik dan modus untuk kolom kategorikal, karena ini adalah metode yang umum untuk mengisi nilai yang hilang yang tidak memiliki informasi yang cukup untuk estimasi yang lebih spesifik.

Langkah ini melibatkan:

Mengisi nilai yang hilang pada kolom numerik (tapOutStopsLat, tapOutStopsLon, stopEndSeq) dengan median.

Mengisi nilai yang hilang pada kolom kategorikal (tapOutStops, tapOutStopsName, tapOutTime) dengan modus (nilai yang paling sering muncul).

```python
# For numerical columns, fill with median
numerical_columns = ['tapOutStopsLat', 'tapOutStopsLon', 'stopEndSeq']
for col in numerical_columns:
    median_value = original_data[col].median()
    original_data[col].fillna(median_value, inplace=True)

# For categorical columns, fill with mode
categorical_columns = ['tapOutStops', 'tapOutStopsName', 'tapOutTime']
for col in categorical_columns:
    mode_value = original_data[col].mode()[0]
    original_data[col].fillna(mode_value, inplace=True)

# Check the remaining missing values for these tapOut columns after filling
original_data[tapout_columns].isna().sum()
```

Sekarang, semua nilai yang hilang pada kolom-kolom yang berkaitan dengan informasi tapOut telah berhasil diisi sepenuhnya.

Kita akan mulai dengan tapInStops terlebih dahulu.

Kita bisa menggunakan variabel lain yang berkaitan erat dengan tapInStops, seperti tapInStopsName, tapInStopsLat, atau tapInStopsLon. Jika kolom ini memiliki informasi yang cukup, kita bisa mengisi tapInStops berdasarkan korelasi langsung. Misalnya, jika sebuah baris memiliki latitude dan longitude yang sangat dekat dengan halte yang terdaftar, kita bisa mengasumsikan halte terdekat sebagai tapInStops.

Maka dari itu kita akan mengimputasi missing value tersebut dengan kombinasi data yang terkait

```python
# Creating a mapping from the combination of 'tapInStopsName', 'tapInStopsLat', and 'tapInStopsLon' to 'tapInStops'
tap_in_stops_mapping = original_data.dropna(subset=['tapInStops']).drop_duplicates(
    subset=['tapInStopsName', 'tapInStopsLat', 'tapInStopsLon']).set_index(
    ['tapInStopsName', 'tapInStopsLat', 'tapInStopsLon'])['tapInStops'].to_dict()

# Function to fill missing 'tapInStops' based on the mapping
def fill_tap_in_stops(row):
    if pd.isna(row['tapInStops']):
        return tap_in_stops_mapping.get((row['tapInStopsName'], row['tapInStopsLat'], row['tapInStopsLon']))
    return row['tapInStops']

# Apply the function to fill missing 'tapInStops'
original_data['tapInStops'] = original_data.apply(fill_tap_in_stops, axis=1)

# Check the remaining missing values for 'tapInStops' after filling
original_data['tapInStops'].isna().sum()
```

Penurunan ini menunjukkan bahwa pendekatan ini cukup efektif untuk sebagian besar data, namun masih ada beberapa baris yang tidak memiliki informasi yang cukup dalam variabel terkait untuk melakukan imputasi. Untuk 34 entri yang masih memiliki nilai yang hilang, kita bisa mempertimbangkan untuk gunakan estimasi jarak yang paling sesuai dengan tempuh perjalanan mereka dilihat dari tapoutstops keluar nya

Untuk mengisi 34 value yang hilang pada kolom `tapInStops` berdasarkan estimasi jarak terdekat dengan `tapOutStops`, kita bisa mengambil pendekatan di mana kita mencari halte yang memiliki jarak geografis terdekat ke `tapOutStops` yang diketahui. Langkah ini akan menggunakan informasi latitude dan longitude untuk menghitung jarak dan memilih halte masuk yang paling logis berdasarkan kedekatan fisik.

### Langkah-langkah:
1. **Ekstraksi Data**: Kumpulkan informasi latitude dan longitude dari semua halte yang tercatat dalam dataset.
2. **Perhitungan Jarak**: Hitung jarak antara lokasi `tapOutStops` dan semua halte lain yang memiliki data lengkap pada `tapInStops`.
3. **Pemilihan Halte**: Pilih `tapInStops` yang memiliki jarak terdekat ke `tapOutStops` sebagai nilai yang diimputasi.



```python
from scipy.spatial.distance import cdist

# Extract known stops with coordinates
known_stops = original_data.dropna(subset=['tapInStops', 'tapInStopsLat', 'tapInStopsLon'])
coordinates = known_stops[['tapInStops', 'tapInStopsLat', 'tapInStopsLon']].drop_duplicates()

# For missing 'tapInStops', calculate the closest stop based on 'tapOutStopsLat' and 'tapOutStopsLon'
def find_closest_stop(row):
    if pd.isna(row['tapInStops']):
        # Calculate distances to all known stops
        distances = cdist([(row['tapOutStopsLat'], row['tapOutStopsLon'])],
                          coordinates[['tapInStopsLat', 'tapInStopsLon']],
                          metric='euclidean')
        # Find the index of the minimum distance
        min_index = distances.argmin()
        return coordinates.iloc[min_index]['tapInStops']
    return row['tapInStops']

# Apply the function to fill missing 'tapInStops'
original_data['tapInStops'] = original_data.apply(find_closest_stop, axis=1)

# Verify the filling
original_data['tapInStops']
```

Kita telah berhasil mengisi 34 nilai yang hilang pada kolom `tapInStops` dengan menggunakan pendekatan berbasis jarak. Untuk setiap nilai yang hilang, saya telah memilih halte masuk (`tapInStops`) yang paling dekat secara geografis dengan halte keluar (`tapOutStops`) berdasarkan koordinat latitude dan longitude.

Output diatas adalah daftar halte yang telah ditetapkan sebagai `tapInStops` berdasarkan kedekatan dengan halte keluar mereka. Anda dapat melihat bahwa setiap entri sekarang memiliki halte yang diimputasi, yang seharusnya mencerminkan lokasi masuk yang paling logis berdasarkan data perjalanan yang tersedia.

- **Kampanye berbasis usia**: Menyasar kelompok usia dengan promosi yang sesuai, seperti penawaran khusus bagi pemuda dan lansia untuk meningkatkan adopsi dan loyalitas. Dan juga diskon harga pada kelompok usia terbanyak (25-40)
- **Personalisasi berdasarkan data perjalanan**: Menawarkan promosi pada waktu dan lokasi yang paling sering dikunjungi oleh segmen tertentu. (Daerah sekolah ditargetkan pada promosi untuk usia dibawah 25 tahun, Daerah perindustrian ditargetkan promosi untuk usia 25 keatas, Daerah tempat hiburan ditargetkan promosi untuk semua usia secara merata)


### 3. Optimalisasi Jadwal dan Lokasi Promosi
**Analisis**:
- Kegiatan pengguna paling tinggi pada jam berangkat dan pulang kerja serta hari kerja, khususnya Jumat.

**Tindakan**:
- **Promosi di waktu puncak**: Melakukan aktivitas promosi intensif pada jam sibuk di halte-halte dengan lalu lintas tinggi. (Senin-Jumat pada Jam 6-7 pagi dan 5-6 sore)
- **Promosi geografis terfokus**: Menggunakan data halte yang paling sering dikunjungi untuk melakukan aktivitas marketing yang lokasional. (Penjaringan, Garuda Taman Mini, dan BKN)
